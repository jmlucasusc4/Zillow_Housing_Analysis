### Principle Component Analysis

```python
###Preliminary Data Exploration

dataset.columns
```

```python
dataset.info()
```

```python
dataset.head()
```

```python
dataset.shape
```

```python
#Extract Numerical Columns
dataset_numerical_features = [col for col in dataset.columns if dataset[col].dtype == 'int64'] 

dataset_numerical_features
```

```python
# extract the numerical only dataset
numerical_features_dataset = dataset[dataset_numerical_features]
```

```python
numerical_features_dataset.head()
```

```python
#Check for missing values
numerical_features_dataset.isnull().sum()
```

```python
for col in dataset_numerical_features:
    numerical_features_dataset[col] = numerical_features_dataset[col].fillna(numerical_features_dataset[col].median())
```

```python
numerical_features_dataset.isnull().sum()
```

```python
#Scale the data
from sklearn.preprocessing import MinMaxScaler

min_max_scaler = MinMaxScaler()

min_max_scaler.fit(numerical_features_dataset)
```

```python
# get scaled data 
dataset_scaled = pd.DataFrame(min_max_scaler.transform(numerical_features_dataset))

# assign column labels
dataset_scaled.columns = dataset_numerical_features
```

```python
dataset_scaled.describe()
```

```python
###PCA
# import PCA from decomposition
from sklearn.decomposition import PCA

# create the PCA object to use to fit the given data
pca = PCA()

# fit the PCA with the given cleaned-up data
pca.fit(dataset_scaled)
```

```python
# Principal axes in feature space, representing the directions of maximum variance in the data.
print(pca.components_)
```

```python
# check the shape 
print(pca.components_.shape)
```

```python
# check the type 
print(type(pca.components_))
```

```python
# Percentage of variance explained by each of the selected components
print(pca.explained_variance_)
```

```python
print(len(pca.explained_variance_))
```

```python
print(pca.singular_values_)
```

```python
print(pca.mean_)
```

```python
print(pca.noise_variance_)
```

```python
principal_components = pca.fit_transform(dataset_scaled)

print(type(principal_components))
print(principal_components.shape)

principal_components_df = pd.DataFrame(principal_components)
```

```python
import seaborn as sns

sns.heatmap(principal_components_df.corr())
```

```python
#Choosing numbers of PCA
import matplotlib.pyplot as plt

plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('number of components')
plt.ylabel('cumulative explained variance');
```

###Linear Regression

```python
from sklearn.model_selection import train_test_split
from sklearn import preprocessing
from sklearn import linear_model
from sklearn.metrics import r2_score,mean_squared_error
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import Lasso
```

```python
dataset_scaled.head()
```

```python
x= dataset_scaled.drop(['SalePrice'],axis=1)
y= dataset_scaled['SalePrice']
```

```python
x.head()
```

```python
y.head()
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20)
```

```python
#Train the model
linear_regression_model=linear_model.LinearRegression()
dataset_num=linear_regression_model.fit(x_train,y_train)
```

```python
#Test the model
```

```python
pred_train=dataset_num.predict(x_train)
pred_test=dataset_num.predict(x_test)
```

```python
#Check Accuracy of R2 Scores
train_set_r2 = r2_score(y_train, pred_train)
print('Linear model, R2 train score is : {}'.format(train_set_r2))
test_set_r2 = r2_score(y_test, pred_test)
print('Linear model, R2 test score is : {}'.format(test_set_r2))
```

```python
# Scatter plot of experimental vs. predicted Logs
import matplotlib.pyplot as plt
import numpy as np
plt.figure(figsize=(5,11))
plt.subplot(2,1,1)
plt.scatter(x=y_train,y=pred_train, c='#7CAE00',alpha=0.3)
z=np.polyfit(y_train,pred_train,1)
p = np.poly1d(z)
plt.plot(y_test,p(y_test),'#F8766D')
plt.ylabel('Predicted Logs')

z = np.polyfit(y_train, pred_train, 1)
p = np.poly1d(z)
plt.plot(y_test,p(y_test),"#F8766D")

plt.ylabel('Predicted LogS')

plt.subplot(2, 1, 2)
plt.scatter(x=y_test, y=pred_test, c="#619CFF", alpha=0.3)

z = np.polyfit(y_test, pred_test, 1)
p = np.poly1d(z)
plt.plot(y_test,p(y_test),"#F8766D")

plt.ylabel('Predicted LogS')
plt.xlabel('Experimental LogS')
plt.show()
```
